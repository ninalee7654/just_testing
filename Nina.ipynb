Concat 2014/06 prices and save as "Result.csv"

```python
import pandas as pd
import glob
import os
path = r'/content/https:/Tankerdata/prices/2014/06' # use the path
all_files = glob.glob(path + "/*.csv")
li = []
for filename in all_files:
    df = pd.read_csv(filename, index_col=None, header=0)
    li.append(df)
data= pd.concat(li, axis=0, ignore_index=True)
data.to_csv('Result.csv')                 #save as a new csv file, named Result 
test= pd.read_csv("Result.csv")
test.sample(20)
```

Combine folders: 
copy 2014/08 to 2014/07, then del 2014/08

```python
import os,shutil
from tqdm import tqdm  
sourceDir=r"/content/https:/Tankerdata/prices/2014/08",
targetDir=r"/content/https:/Tankerdata/prices/2014/07"
for root, dirs, files in os.walk(sourceDir):
    for file in tqdm(files,desc="Waiting..."):
        shutil.copy(os.path.join(root,file),targetDir) 
shutil.rmtree(sourceDir)
```
